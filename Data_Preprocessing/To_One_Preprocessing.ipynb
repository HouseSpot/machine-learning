## A Little Bit of Preprocessing

```python
df_99co_bogor_bagas[['City']] = df_99co_bogor_bagas[['City']].applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
df_99co_bogor_bagas
```

```python
df_99co_bogor_bagas['City'] = df_99co_bogor_bagas['City'].replace({"Kabupaten Bogor": "Bogor"})
```

```python
df_99co_bogor_bagas
```

## To One

```python
# List of DataFrames
dfs = [
    df_99co_bekasi_nopal,
    df_99co_bogor_nopal,
    df_99co_tangerang_nopal,
    df_99co_depok_nopal,
    df_99co_bogor_bagas,
    df_99co_jakarta_bagas,
    df_rumah123_jakarta_salma
]

# Combine all DataFrames into one DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

combined_df
```

```python
combined_df.info()
```

```python
unique_counts = combined_df['City'].value_counts()

unique_counts
```

```python
# Create a mask to identify rows where 'city' is 'Banten', 'Jakarta', or 'Jawa Barat'
mask = combined_df['City'].isin(['Banten', 'Jakarta', 'Jawa Barat'])

# Drop rows where 'city' is 'Banten', 'Jakarta', or 'Jawa Barat'
combined_df = combined_df[~mask]
```

```python
combined_df['City'] = combined_df['City'].replace({'Kota Bogor': 'Bogor'})
```

```python
# Update 'city' and 'district' columns based on conditions
mask = combined_df['City'] == 'Kali Deres'
combined_df.loc[mask, 'City'] = 'Jakarta Barat'
combined_df.loc[mask, 'District'] = 'Kali Deres'
```

```python
# Update 'city' and 'district' columns based on conditions
mask = combined_df['City'] == 'Setia Budi'
combined_df.loc[mask, 'City'] = 'Jakarta Selatan'
combined_df.loc[mask, 'District'] = 'Setia Budi'
```

```python
unique_counts = combined_df['City'].value_counts()

unique_counts
```

```python
combined_df
```

```python
combined_df = combined_df.drop_duplicates()
```

```python
combined_df
```

```python
unique_counts = combined_df['City'].value_counts()

unique_counts
```

## Sanity Check

```python
# Compute maximum and minimum values for each column
max_values = combined_df.max()
min_values = combined_df.min()

# Construct dictionary with maximum and minimum values for each column
result_dict = {}
for column in combined_df.columns:
  if(column == "Price"):
    continue
  result_dict[f"{column}_max"] = max_values[column]
  result_dict[f"{column}_min"] = min_values[column]

result_dict
```

```python
combined_df.drop(combined_df[combined_df['Building_Size_M2'] < 20].index, inplace=True)
combined_df.drop(combined_df[combined_df['Bathrooms'] < 0].index, inplace=True)
combined_df.drop(combined_df[combined_df['Land_Size_M2'] < 20].index, inplace=True)

combined_df
```

```python
# Compute maximum and minimum values for each column
max_values = combined_df.max()
min_values = combined_df.min()

# Construct dictionary with maximum and minimum values for each column
result_dict = {}
for column in combined_df.columns:
  if(column == "Price"):
    continue
  result_dict[f"{column}_max"] = max_values[column]
  result_dict[f"{column}_min"] = min_values[column]

result_dict
```

```python
combined_df['City'] = combined_df['City'].str.title()
combined_df['District'] = combined_df['District'].str.title()
```
