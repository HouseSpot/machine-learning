## Null Values Handler

```python
def nullCheck(df):
    # Get the number of null values in each column
    null_counts = df.isnull().sum()

    # Print the column names and their corresponding number of null values
    for column, null_count in null_counts.items():
        print(f"Column: {column}, Null values: {null_count}")
```

```python
nullCheck(df)
```

```python
df[['Carport','Garasi']] = df[['Carport','Garasi']].fillna(0)
```

```python
nullCheck(df)
```

```python
df = df.dropna()
```

```python
nullCheck(df)
```

```python
df.info()
```

## Data Type Transformer

```python
def find_non_float_rows(df, features):
    # Function to check if a value can be converted to float
    def can_be_float(value):
        try:
            float(value)
            return True
        except ValueError:
            return False

    # Apply the function across the specified features
    mask = df[features].applymap(can_be_float)

    # Identify rows that contain any non-float values in the specified features
    non_float_rows = df[~mask.all(axis=1)]

    return non_float_rows
```

```python
# List of features (columns) to check
features = ['latitude', 'longitude', 'price_value']

# Find rows with non-float values in the specified features
non_float_rows = find_non_float_rows(df, features)

non_float_rows
```

```python
df['latitude'] = df['latitude'].astype(float)
df['longitude'] = df['longitude'].astype(float)
```

```python
# List of features (columns) to check
features = ['price_value']

# Find rows with non-float values in the specified features
non_float_rows = find_non_float_rows(df, features)

non_float_rows
```

```python
# Convert all columns to numeric, coerce errors to NaN
df = df[df['price_value'] != 'unknown']

# Drop rows with any NaN values
df
```

```python
# List of features (columns) to check
features = ['price_value']

# Find rows with non-float values in the specified features
non_float_rows = find_non_float_rows(df, features)

non_float_rows
```

```python
df
```

```python
# Replace commas with periods
df['price_value'] = df['price_value'].str.replace(',', '.')

# Convert the column to float
df['price_value'] = df['price_value'].astype(float)

df
```

```python
# List of features (columns) to check
features = ['price_value']

# Find rows with non-float values in the specified features
non_float_rows = find_non_float_rows(df, features)

non_float_rows
```

```python
df.info()
```

```python
df['Luas Bangunan'] = df['Luas Bangunan'].str.extract(r'(\d+)').astype(float)
df['Luas Tanah'] = df['Luas Tanah'].str.extract(r'(\d+)').astype(float)

# List of features (columns) to check
features = ['Luas Bangunan','Luas Tanah']

# Find rows with non-float values in the specified features
non_float_rows = find_non_float_rows(df, features)

non_float_rows
```

```python
df.info()
```

```python
df['price_currency'].unique()
```

```python
df['price_unit'].unique()
```

```python
# Define conversion factors
conversion_factors = {
    'Miliar': 10**9,  # 1 Milliar is 1 billion
    'Juta': 10**6      # 1 Juta is 1 million
}

# Function to calculate price
def calculate_price(row):
    return row['price_value'] * conversion_factors[row['price_unit']]

# Apply the function to create the new 'price' column
df['price'] = df.apply(calculate_price, axis=1)

df
```

```python
df.info()
```

```python
# List of features to drop
features_to_drop = ['price_currency', 'price_value','price_unit']

# Drop the features
df = df.drop(features_to_drop, axis=1)

df.info()
```

## Split Address into District and City

```python
# Get unique counts of commas in each value of the 'feature_with_commas' column
comma_counts = df['address'].str.count(',')

# Print unique counts
comma_counts.unique()
```

```python
# Split the 'location' column on comma and expand into separate columns
split_data = df['address'].str.split(',', expand=True)

# Assign the split parts to new columns
df['District'] = split_data[0].str.strip()
df['City'] = split_data[1].str.strip()

# List of features to drop
features_to_drop = ['address']

# Drop the features
df = df.drop(features_to_drop, axis=1)

df.info()
```

```python
df
```

## Garage and Carport Combine Feature

```python
# Calculate the maximum of feature1 and feature2 across rows
df['Garage/Carport'] = df[['Garasi', 'Carport']].max(axis=1)

# List of features to drop
features_to_drop = ['Garasi','Carport']

# Drop the features
df = df.drop(features_to_drop, axis=1)

df
```

## Rename Feature and Save
